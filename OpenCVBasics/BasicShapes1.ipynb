```python
import cv2 
import numpy as np 
```

```python
import matplotlib.pyplot as plt
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16) # data type = int 16 bit
```

```python
# blank_img
plt.imshow(blank_img)
```

```python
rect=cv2.rectangle(blank_img,pt1=(400,0),pt2=(510,150),color=(0,240,0),thickness=10)
```

```python
plt.imshow(rect)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=10)
```

```python
plt.imshow(blank_img)
```

### Circle 

```python
cv2.circle(img=blank_img,center=(100,100),radius=50,color=(255,0,0),thickness=9)
plt.imshow(blank_img)
```

```python
cv2.circle(img=blank_img,center=(400,400),radius=50,color=(255,0,0),thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,pt1=(0,70),pt2=(512,512),color=(102,255,255),thickness=5)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='Hello',org=(10,500),fontFace=font,fontScale=4,color=(255,255,255),thickness=2,lineType=cv2.LINE_AA)
plt.imshow(blank_img)
```

##### Drawing Polygon

```python
vertices = np.array([  [100,300], [200,200],[400,300] , [200,400] ],dtype=np.int32)
```

```python
vertices
```

```python
vertices.shape
```

```python
pts=vertices.reshape((-1,1,2))
```

```python
pts.shape
```

```python
cv2.polylines(blank_img2,[pts],isClosed=True,color=(255,0,0),thickness=5)
```

```python
plt.imshow(blank_img2)
```
